```python
# Step 1: Import pandas

# pandas is used to create and manage data in table format
import pandas as pd
```

```python
# Step 2: Create the dataset

# This dataset shows:
# user    -> who gave the rating
# product -> which product was rated
# rating  -> how much the user liked the product

data = {
    "user": ["Amit", "Amit", "Riya", "Riya", "Sohan", "Sohan", "Neha", "Neha", "Kiran", "Kiran"],
    "product": ["AI Book", "Python Book", "AI Book", "Marketing Book", "Python Book", "Finance Book", "AI Book", "Finance Book", "Marketing Book", "AI Book"],
    "rating": [5, 4, 4, 5, 5, 3, 4, 4, 5, 5]
}

# Convert dictionary into DataFrame
df = pd.DataFrame(data)

# Display dataset
df
```

```python
# Step 3: Inspect the data

# head() shows the first 5 rows
print("First 5 rows:")
print(df.head())

# shape shows number of rows and columns
print("\nShape of dataset:")
print(df.shape)

# info() shows column names, data types, and non-null values
print("\nDataset information:")
print(df.info())
```

```python
# Step 4: Understand user-product interactions

# Each row represents one interaction between a user and a product
# This helps businesses understand what users like and dislike

print("User-Product Rating Data:")
print(df)
```

```python
# Step 5: Find average rating of each product

# groupby("product") groups all rows product-wise
# ["rating"].mean() calculates average rating for each product
# sort_values(ascending=False) sorts products from highest to lowest rating

product_ratings = df.groupby("product")["rating"].mean().sort_values(ascending=False)

# Display average rating by product
print("Average Rating of Each Product:")
print(product_ratings)
```

```python
# Step 6: Convert the result into a DataFrame for better display

# reset_index() converts the grouped result into a normal table
recommendations = product_ratings.reset_index()

# Rename column names for better presentation
recommendations.columns = ["Product", "Average Rating"]

# Display recommendation table
recommendations
```

```python
# Step 7: Show top recommended products

# head(3) gives the top 3 most highly rated products
top_recommendations = recommendations.head(3)

print("Top Recommended Products:")
print(top_recommendations)
```

```python
# Step 8: Business interpretation

# These top-rated products can be:
# 1. shown on the homepage
# 2. recommended to new users
# 3. promoted in email campaigns
# 4. used for upselling and cross-selling

print("Business Interpretation:")
print("The products shown above have the highest average ratings.")
print("These can be recommended first to improve customer engagement and sales.")
```

```python
# Step 9: Full recommendation engine result in one view

print("Complete Recommendation Table:")
print(recommendations)

print("\nTop 3 Recommended Products:")
print(top_recommendations)
```

“This is a popularity-based recommendation engine. It checks which products received the highest average ratings and recommends those products first. In real companies like Amazon, Netflix, and YouTube, the logic becomes more advanced, but the purpose remains the same: show relevant items to users.”
